### A7

Pasinaudokite A6 užduoties kodu.

Parašykite funkciją, kuri priimtų du argumentus `x` verčių sąrašą bei `k` - konstantą.

Funkcija turėtų atvaizduoti $\sin(k\pi x)$ grafiką.

Pvz. `grafikas(x, 3)` turėtų atvaizduoti $\sin(3\pi x)$ kreivę.

````python
import math
import matplotlib.pyplot as plt


X = []
y_sin = []
y_cos = []

def grafikas(x, k):
  for i in range(x):
    temp = i * 0.01
    X.append(temp)
    y_sin.append(math.sin(math.pi * temp * k))
    y_cos.append(math.cos(2 * math.pi * temp * k))

x0 = int(input("Įveskite norimą verčių kiekį: "))
k0 = int(input("Įveskite norimą k konstantą: "))

grafikas(x0, k0)

fig, ax = plt.subplots()
ax.plot(X, y_sin, color="black", lw=2, label=f"y = sin({k0}πx)")

ax.set(title=f"Grafikas: sin({k0}πx)",
       xlabel="x reikšmės",
       ylabel="y reikšmės")
ax.legend()
fig.suptitle("3 paveikslėlis")

plt.show()
````

### B1

[Absoliučiai juodas kūnas](https://en.wikipedia.org/wiki/Black_body) - fizikinė abstrakcija, reiškianti kūną, kuris sugeria visas į jį krentančias elektromagnetines bangas, t.y. neatsipindi jokių elektromagnetinių bangų.
Nepaisant to, absoliučiai juodas kūnas gali spinduliuoti elektromagnetines bangas ir turėti savo spalvą. Jo spinduliavimo spektrą apsprendžia vienintelis parametras - temperatūra. Dėl to absoliučiai juodas kūnas kartais vadinamas idealiuoju spinduoliu.

Terminą "absoliučiai juodas kūnas" 1860 m. pirmą kartą panaudojo Gustavas Kirchhofas. Maksas Plankas 1900 m. empiriškai išvedė absoliučiai juodo kūno [spinduliuotės dėsnį](https://en.wikipedia.org/wiki/Planck\%27s_law):
\begin{equation}
B(\lambda,T) = \frac{2 h c^2}{\lambda^5} \cdot \frac{1}{\exp\bigg(\dfrac{hc}{\lambda k_B T}\bigg)-1}
\end{equation}
kur $B(\lambda,T)$ yra energijos kiekis iš ploto vieneto per laiko vienetą į erdvinio kampo vienetą, spinduliuojamas dažnių intervale tarp $\lambda$ ir $\lambda+d\lambda$,

$T$ - absoliučiai juodo kūno temperatūra,

$h$ - Planko konstanta,

$c$ - šviesos greitis vakuume,

$k_B$ - Bolcmano konstanta,

$e$ - Eulerio konstanta.
