[Rydbergo konstanta](https://en.wikipedia.org/wiki/Rydberg_constant), pavadinta švedų fiziko Johano Rydbergo vardu, rodo bet kurio fotono, kurį gali išspinduliuoti atomas, didžiausią bangos skaičių (atvirkštinį bangos ilgį), arba atvirkščiai - mažiausią bangos skaičių, galintį jonizuoti  atomą.  

Paprastumo dėlei, laikant, kad atomo branduolio masė yra begalinė elektrono masės atžvilgiu, Rydbergo konstanta yra

$R_{\infty} = \frac{m_e e^4}{8 \varepsilon_0^2 h^3 c} =  10973731.568160(21) \text{m}^{-1}$, kur

$h$ yra Planko konstanta

$m_e$ - elektrono masė

$e$ - elementarusis krūvis

$c$ - šviesos greitis vakuume

$\varepsilon_0$ - vakuumo dielektrinė skvarba.
